```python
from collections import deque
from sklearn.preprocessing import StandardScaler
import numpy as np
from tqdm import tqdm
import pandas as pd
import torch
```

```python
"""
look_up : window for calculating the direction of price evolution
alpha : alpha treshold from the paper 
lookback : boolean to indicate if looking back for calculating labels
normalization_window : number of samples in the past for normalizing
prediction_window : number of samples as input of the model
"""

params = {
    'look_up' : 1024,     
    'alpha' : 0.1,
    'look_back' : False,
    'normalization_window' : 10000,
    'prediction_window' : 512,
    'channels_conv_inception' : 16,
    'channels_inception_lstm' : 32,
    'hidden_dim_lstm' : 64,
    'n_classes' : 3,
    'batch_size': 256
}
```

```python

class DatasetPreprocessing :
    """
    Parameters :
    k : window for calculating the direction of price evolution
    nw : number of samples for normalization
    tw : number of samples for input of the model
    a : alpha treshold from the paper 
    """
    def __init__ (self,k = params['look_up'], a = params['alpha'],look_back = params['look_back'],
                  nw=params['normalization_window'],tw=params['prediction_window'],path = 'public_quote_bitmex_XBTUSD.csv'):
       
        self.dataset = None
        self.datasetNormalized = None
        self.n = None
        self.l = None
        self.i = None
        self.path = path
        self.look_back = False
        
        self.k = k
        self.nw = nw
        self.tw = tw
        self.a = a

    """
    Data access throw the unprocessed dataset
    Could easily be changed in lazy loading approach / stream approach
    """


    def loadDataset(self):
        self.dataset = pd.read_csv(self.path)
        self.initial_dataset = self.dataset.copy()
        self.dataset = self.dataset.drop('time',axis=1)
        self.n,self.l = self.dataset.shape
        self.sum = np.zeros(self.l-1)

    def reinitializeDataset(self):
        self.i = len(self.dataset)-1
        
    def getNextRow(self):
        self.i -= 1
        return np.array(self.dataset.loc[self.i+1,:],dtype = np.float64)
    
        

    """
    Actual processing
    """ 
    
    def normalizeDataset(self):
        for column in self.dataset.columns :
            print(f"column {column} being normalized...")
            feature = self.dataset.loc[:,column]
            
            means = feature.rolling(self.nw).mean()
            stds = feature.rolling(self.nw).std()
            
            feature = (feature - means) / stds
            self.dataset.loc[:,column] = feature
    
    def labellizeDataset(self):
        self.dataset['mid_price'] = (self.dataset['ask_price']+self.dataset['bid_price'])/2
        mid_price = self.dataset['mid_price']
        back_price = mid_price.rolling(self.k).mean()
        head_price = mid_price.rolling(self.k).mean().shift(-self.k)
        if(self.look_back):
            self.dataset['target'] =  (head_price - back_price) #/ back_price
        else: 
            self.dataset['target'] =  (head_price - mid_price) #/ mid_price
        self.dataset['target'] = self.dataset['target'].apply(self.UpOrDown)
            

        
    def UpOrDown(self,value):
        return (abs(value)>self.a)*((-1)**(value<-self.a))
    
    """
    Access data throw the processed Dataset
    """
    
    def get(self):
        self.loadDataset()
        self.normalizeDataset()
        self.labellizeDataset()
        return self.dataset.iloc[self.nw:-self.k]
        
        
        
        
```

# Data Loader

```python
from torch.utils.data import Dataset, DataLoader

class OrderBookDataset(Dataset):
    def __init__(self,train,timestamp = params['prediction_window']):
        self.parameters = DatasetPreprocessing(1000,0.1)
        self.dataset = self.parameters.get()
        if(train == 'train'):
            n = int(len(self.dataset)*0.6)
            self.dataset = self.dataset.iloc[:n]
        elif(train == 'validation'):
            n1 = int(len(self.dataset)*0.6)
            n2 = int(len(self.dataset)*0.8)
            self.dataset = self.dataset.iloc[n1:n2]
        else:
            n = int(len(self.dataset)*0.8)
            self.dataset = self.dataset.iloc[n:]
        self.timestamp = timestamp
        
    def __len__(self):
        return len(self.dataset)-self.timestamp
    
    def __getitem__(self,idx):
        start_idx = self.dataset.index[0]
        #print(self.dataset[['ask_price','ask_volume','bid_price','bid_volume']].loc[start_idx+idx:start_idx+idx+self.timestamp])
        #print(self.timestamp)
        X = self.dataset[['ask_price','ask_volume','bid_price','bid_volume']].loc[start_idx+idx+1:start_idx+idx+self.timestamp]
        y = self.dataset.loc[start_idx+idx+self.timestamp,'target']+1
        return torch.tensor(np.array(X)[np.newaxis]).type('torch.FloatTensor'),y
       
        
```

```python
import torch.nn as nn
import torch.nn.functional as F

class ConvNet(nn.Module):
    def __init__(self,timestamp = params['prediction_window'],out_channels = params['channels_conv_inception'],lr_parameter = 0.01):
        super().__init__()
        self.lr_parameter = lr_parameter
        self.timestamp = timestamp
        self.conv = nn.Conv2d(in_channels=1,out_channels=out_channels,kernel_size=(1,2),stride=(1,2))
        self.conv2 = nn.Conv2d(in_channels=out_channels,out_channels=out_channels,kernel_size=(5,1),padding = (2,0))
        self.conv3 = nn.Conv2d(in_channels=out_channels,out_channels=out_channels,kernel_size=(1,2),stride=(1,2))

    def forward(self, x):
        #print(x.shape)
        lr = nn.LeakyReLU(self.lr_parameter)
        x = lr(self.conv(x))
        #print(x.shape)
        x = lr(self.conv2(x))
        #print(x.shape)
        x = lr(self.conv2(x))
        #print(x.shape)
        x = lr(self.conv2(lr(self.conv2(lr(self.conv3(x))))))
        #print(x.shape)
        return x


net = ConvNet()
from torchsummary import summary
summary(net,(1,params['prediction_window'],4))
```

```python
class BasicConv2d(nn.Module):

    def __init__(self, in_channels, out_channels, **kwargs):
        super(BasicConv2d, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, bias=False, **kwargs)
        #self.bn = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        x = self.conv(x)
        #x = self.bn(x)
        return F.relu(x, inplace=True)


class InceptionBasicBlock(nn.Module):

    def __init__(self, in_channels = params['channels_conv_inception'],out_channels = params['channels_inception_lstm']):
        super().__init__()
        
        self.branch5x5_1 = BasicConv2d(in_channels, out_channels, kernel_size=1)
        self.branch5x5_2 = BasicConv2d(out_channels, out_channels, kernel_size=5, padding=2)

        self.branch3x3dbl_1 = BasicConv2d(in_channels, out_channels, kernel_size=1)
        self.branch3x3dbl_2 = BasicConv2d(out_channels, out_channels, kernel_size=3, padding=1)

        self.branch_pool = BasicConv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        
        branch5x5 = self.branch5x5_1(x)
        branch5x5 = self.branch5x5_2(branch5x5)

        branch3x3dbl = self.branch3x3dbl_1(x)
        branch3x3dbl = self.branch3x3dbl_2(branch3x3dbl)

        branch_pool = F.max_pool2d(x, kernel_size=3, stride=1, padding=1)
        branch_pool = self.branch_pool(branch_pool)

        outputs = [branch5x5, branch3x3dbl, branch_pool]
        return torch.cat(outputs, 1)

inception = InceptionBasicBlock()

from torchsummary import summary
summary(inception,(params['channels_conv_inception'],params['prediction_window'],1))
```

```python
class LSTM(nn.Module):

    def __init__(self, input_dim = 3*params['channels_inception_lstm'], hidden_dim = params['hidden_dim_lstm'], batch_size = params['batch_size'], output_dim=params['n_classes'],num_layers=1):
        super(LSTM, self).__init__()
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.batch_size = batch_size
        self.num_layers = num_layers

        # Define the LSTM layer
        self.lstm = nn.LSTM(self.input_dim, self.hidden_dim, self.num_layers)

        # Define the output layer
        self.linear = nn.Linear(self.hidden_dim, output_dim)

    def init_hidden(self):
        # This is what we'll initialise our hidden state as
        return (torch.zeros(self.num_layers, self.batch_size, self.hidden_dim),
                torch.zeros(self.num_layers, self.batch_size, self.hidden_dim))

    def forward(self, inp):
        # Forward pass through LSTM layer
        # shape of lstm_out: [input_size, batch_size, hidden_dim]
        # shape of self.hidden: (a, b), where a and b both 
        # have shape (num_layers, batch_size, hidden_dim).
        #print(inp.shape)
        inp = inp.permute(2,0,1,3)
        inp = torch.squeeze(inp)
        print(inp.shape)
        print(inp.view(len(inp), self.batch_size, -1).shape)
        lstm_out, self.hidden = self.lstm(inp.view(len(inp), self.batch_size, -1))
        
        # Only take the output from the final timetep
        # Can pass on the entirety of lstm_out to the next layer if it is a seq2seq prediction
        y_pred = self.linear(lstm_out[-1].view(self.batch_size, -1))
        return y_pred

model = LSTM()
#summary(model,(48,256,1))
```

# Training

```python
from torch.optim import Adam
import time as t
from tqdm import tqdm

#params
params_training = {
    'batch_size': params['batch_size'],
    'shuffle': True,
    'num_workers': 4,
    'lr' : 0.001
    }
max_epochs = 50



use_cuda = torch.cuda.is_available()
device = torch.device("cuda:0" if use_cuda else "cpu")

net = full
net = net.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = Adam(net.parameters(), lr=params_training['lr'])
```

```python
# Generators
training_set = OrderBookDataset('train')
training_generator = DataLoader(training_set, **params_training)

validation_set = OrderBookDataset('validation')
validation_generator = DataLoader(validation_set, **params_training)


for epoch in range(max_epochs):
    # Training
    pbar = tqdm(total = params['batch_size']//training_set.__len__()+1)
    t1 = t.time()
    for inputs, labels in training_generator:
        inputs,labels = inputs.to(device),labels.to(device)
        optimizer.zero_grad()
        outputs = net(inputs)
        #print(outputs.shape,labels.shape)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        pbar.update()
    
    print(f"time elapsed : {t.time()-t1} seconds")
    # Validation
    with torch.set_grad_enabled(False):
        validationLoss = 0
        counter = 0
        for inputs, labels in validation_generator:
            inputs,labels = inputs.to(device),labels.to(device)
            outputs = net(inputs)
            validationLoss += criterion(outputs, labels)
            counter +=1
        print(f"epoch {epoch} | validation loss : {validationLoss/counter}")
                
```
